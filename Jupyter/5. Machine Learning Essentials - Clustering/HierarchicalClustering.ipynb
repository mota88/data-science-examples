```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

df = pd.read_csv("Z:/csv/base_datos_2008.csv", nrows=1e4)
newdf = df[['AirTime','DepDelay']].dropna()
```

```python
clstr = AgglomerativeClustering(n_clusters=5)
clstr.fit(newdf)

# euclidean / l1 / l2 / manhattan / cosine / precomputed
# affinity='l1', linkage='complete'
```

```python
plt.scatter(newdf['AirTime'],newdf['DepDelay'],c=clstr.fit_predict(newdf))
plt.show()
```
