```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import time
import matplotlib.dates as md
import matplotlib.ticker as ticker

# custom function to convert a string percent to a float
def p2f(string):
    x = string.replace(',', '.')
    return float(x.strip('%'))#/100

sns.set(style="darkgrid")
df = pd.read_csv("Z:/csv/data.csv", parse_dates=['informe','última vacuna registrada'], dayfirst=True, converters={'% sobre entregadas':p2f})
dftotales = df[df['comunidad autónoma'] == 'Totales']
#df['última vacuna registrada'] = pd.to_datetime(df['última vacuna registrada'], dayfirst = True)
df.dtypes
```

```python
df = df[df['última vacuna registrada'].isna() == False]
#df['última vacuna registrada'] = pd.to_datetime(df['última vacuna registrada'], format='%d/%m/%Y', dayfirst=True)
#df.sort_values(by=['última vacuna registrada'], inplace=True, ascending=True)
df.tail(19)
```

```python
data = df.groupby(by=['última vacuna registrada'],as_index=False)['dosis administradas'].sum()
data
```

```python
sns.set(rc={'figure.figsize':(15,8)})
ax = sns.lineplot(data['última vacuna registrada'],data['dosis administradas'], palette='pastel', marker='*', markersize=14)
ax.set_title("Vacunación en España: evolución de dosis administradas", fontsize=20)
ax.set_xlabel("Fecha", fontsize=18)
ax.set_ylabel("Dosis Administradas", fontsize=18)
ax.yaxis.set_major_formatter(ticker.EngFormatter())
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(0, end, 100000))
ax.xaxis.set_major_formatter(md.DateFormatter('%d/%m/%Y'))
ax.xaxis.set_major_locator(md.DayLocator(interval=1))
caption="Datos: Civio"
plt.figtext(0.9, 0.0001, caption, horizontalalignment='right', fontsize=14)
plt.setp(ax.xaxis.get_majorticklabels(), rotation = 45, ha="right")
```

```python
gb_df = pd.DataFrame(df.groupby(['última vacuna registrada','comunidad autónoma'],as_index=False)['dosis administradas','dosis entregadas'].sum())
gb_df
```

```python
data = gb_df.pivot('última vacuna registrada','comunidad autónoma','dosis administradas')
data = data.fillna(0)
data.index = data.index.strftime('%d/%m/%Y')
data
```

```python
sns.set(rc={'figure.figsize':(20,18)})
ax = sns.heatmap(data=data, annot=True, linewidths=.5, cmap='Blues', fmt='g')
ax.set_title("Vacunación en España: dosis administradas por Comunidad Autónoma y fecha", fontsize=20)
ax.set_xlabel("Comunidad Autónoma", fontsize=18)
ax.set_ylabel("Fecha", fontsize=18)
#ax.xaxis.set_major_formatter(md.DateFormatter('%d/%m/%Y'))
#ax.xaxis.set_major_locator(md.DayLocator(interval=1))
caption="Datos: Civio"
plt.figtext(0.9, 0.0001, caption, horizontalalignment='right', fontsize=14)
plt.setp(ax.xaxis.get_majorticklabels(), rotation = 45, ha="right")
```

```python
df = df[df['última vacuna registrada'].isna() == False]
autonomias = [(df['comunidad autónoma'] == 'Andalucía'),
              (df['comunidad autónoma'] == 'Aragón'),
              (df['comunidad autónoma'] == 'Asturias'),
              (df['comunidad autónoma'] == 'Baleares'),
              (df['comunidad autónoma'] == 'Canarias'),
              (df['comunidad autónoma'] == 'Cantabria'),
              (df['comunidad autónoma'] == 'Castilla y Leon'),
              (df['comunidad autónoma'] == 'Castilla La Mancha'),
              (df['comunidad autónoma'] == 'Cataluña'),
              (df['comunidad autónoma'] == 'C. Valenciana'),
              (df['comunidad autónoma'] == 'Extremadura'),
              (df['comunidad autónoma'] == 'Galicia'),
              (df['comunidad autónoma'] == 'La Rioja'),
              (df['comunidad autónoma'] == 'Madrid'),
              (df['comunidad autónoma'] == 'Murcia'),
              (df['comunidad autónoma'] == 'Navarra'),
              (df['comunidad autónoma'] == 'País Vasco'),
              (df['comunidad autónoma'] == 'Ceuta'),
              (df['comunidad autónoma'] == 'Melilla')]
poblaciones = [8464411,1329391,1018784,1171543,2175952,580295,2394918,2045221,7780479,5057353,1063987,2701819,319914,6779888,1511251,661197,2220504,84202,87076]
df['poblacion_ccaa'] = np.select(autonomias, poblaciones)

dosisadmin = df['dosis administradas']
dosisrec = df['dosis entregadas']
dosisadminpob = np.array([],dtype = float)
dosisrecpob = np.array([],dtype = float)
for i, pob in enumerate(df['poblacion_ccaa']):
    #dosisitem = dosisadmin[abs(len(data)-19+i):abs(len(data)-18+i)].item()
    dosisadminitem = dosisadmin[i:i+1].item()
    dosisrecitem = dosisrec[i:i+1].item()
    dosisadminpob = np.append(dosisadminpob,round((dosisadminitem/pob)*100000, 2))
    dosisrecpob = np.append(dosisrecpob,round((dosisrecitem/pob)*100000, 2))
df['dosis_admin_por_100m'] = dosisadminpob
df['dosis_rec_por_100m'] = dosisrecpob
```

```python
data = df.tail(19)

fig, ax = plt.subplots(figsize=(20, 9))
g = data.plot(x='comunidad autónoma', y=['dosis administradas', 'dosis entregadas'], kind='bar', ax=ax)
ax.legend(title='Dosis', loc='upper right', labels=['Administradas', 'Recibidas'], fontsize=16)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha='right')
ax.set_title("Vacunación en España: dosis administradas y recibidas por Comunidad Autónoma", fontsize=20)
ax.set_xlabel("Comunidad Autónoma", fontsize=18)
ax.set_ylabel("Dosis", fontsize=18)
caption="Datos: Civio"
plt.figtext(0.9, 0.0001, caption, horizontalalignment='right', fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 9))
g = data.plot(x='comunidad autónoma', y=['dosis_admin_por_100m', 'dosis_rec_por_100m'], kind='bar', ax=ax)
ax.legend(title='Dosis', loc='upper right', labels=['Administradas', 'Recibidas'], fontsize=16)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha='right')
ax.set_title("Vacunación en España: dosis administradas y recibidas por Comunidad Autónoma por cada 100m habitantes", fontsize=20)
ax.set_xlabel("Comunidad Autónoma", fontsize=18)
ax.set_ylabel("Dosis (100m habitantes)", fontsize=18)
caption="Datos: Civio y datos propios"
plt.figtext(0.9, 0.0001, caption, horizontalalignment='right', fontsize=14)

plt.show()
```

```python
threshold = dftotales[dftotales['comunidad autónoma'] == 'Totales'].tail(1).iloc[0]['% sobre entregadas']

fig, ax = plt.subplots(figsize=(20, 9))
#l = sns.lineplot(data[data['comunidad autónoma'] == 'Totales'],data['% sobre entregadas'], palette='pastel')
g = sns.barplot(x=data['comunidad autónoma'], y=data['% sobre entregadas'], palette='tab20', ax=ax)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha='right')
ax.set_title("Vacunación en España: porcentaje de dosis administradas sobre las recibidas por Comunidad Autónoma", fontsize=20)
ax.set_xlabel("Comunidad Autónoma", fontsize=18)
ax.set_ylabel("Porcentaje", fontsize=18)
# horizontal line indicating the threshold
start, end = ax.get_xlim()
ax.plot([start, end], [threshold, threshold], 'k--')
plt.text(end+0.1, threshold-0.5, "España", fontsize=14)
caption="Datos: Civio"
plt.figtext(0.9, 0.0001, caption, horizontalalignment='right', fontsize=14)

plt.show()
```
