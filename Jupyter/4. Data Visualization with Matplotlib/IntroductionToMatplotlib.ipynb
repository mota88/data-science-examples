```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Z:/csv/casos_diagnostico_provincia.csv", nrows = 1000000)
```

```python
import datetime as dt

df.loc[:,'fecha'] =  pd.to_datetime(df['fecha'],infer_datetime_format=True).dt.date
df.dtypes
```

```python
date = dt.date.today() - dt.timedelta(days=100)
df100dias = df[(df['fecha'] > date) & (df['num_casos'] > 2000)]
df100dias
```

```python
data = np.unique(df100dias.num_casos, return_counts=True)
data
```

```python
plt.pie(x=data[1],labels=data[0])
plt.show()
```

```python
plt.pie(x=data[1],labels=data[0],colors=['Red','Green'])
plt.show()
```

```python
plt.pie(x=data[1],labels=data[0],shadow=True)
plt.show()
```

```python
plt.pie(x=data[1],labels=data[0],startangle=90)
plt.show()
```

```python
plt.pie(x=data[1],labels=data[0],startangle=90,radius=2)
plt.show()
```
