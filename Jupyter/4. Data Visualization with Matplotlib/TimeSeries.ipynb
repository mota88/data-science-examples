```python
import numpy as np
import seaborn as sns
import pandas as pd
import datetime
import time
df  = pd.read_csv("Z:/csv/casos_diagnostico_ccaa.csv")
```

```python
#times = []
#for i in np.arange(len(df)):
    #times.append(datetime.datetime(year = 2008, month = df.loc[i,'Month'], day = df.loc[i,'DayofMonth']))
df['fecha'] = pd.to_datetime(df['fecha'])
data = df.groupby(by=['fecha'],as_index=False)['num_casos'].mean()
data.head()
```

```python
sns.lineplot(data['fecha'],data['num_casos'])
```

```python
data = df.groupby(by=['fecha'])['num_casos','num_casos_prueba_pcr'].mean()
data.head()
```

```python
sns.lineplot(data = data)
```

```python
df2 = df[df['ccaa_iso'].isin(['AN','CT','MD','VC'])]
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='fecha',y='num_casos',hue='ccaa_iso',data=df2)
```
